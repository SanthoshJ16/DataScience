# Independant Sample - Unpaired T -Test

```python
## Unpaired T test
## Different Group #Same condition 

from scipy.stats import ttest_ind
male = dataset[dataset['gender']=='M']['salary']
female = dataset[dataset['gender']=='F']['salary']
ttest_ind(male, female)
```

# Dependant Sample - Paired T test

```python
## Paired T test
## Same Group #different condition 
dataset.dropna()
from scipy.stats import ttest_ind
male = dataset[dataset['gender']=='M']['hsc_p']
male1 = dataset[dataset['gender']=='M']['ssc_p']
ttest_ind(male, male1)
```

### Hypothesis Testing 

```python
## Problem Statement : Test whether the ssc_p pass mark of male and hsc_p of Male is same
```

```python
## Null Hypothesis : There is no similarity between ssc_p pass mark of male and hsc_p of Male
```

```python
## From the above Paired T test results , we got a pvalue=0.645 which is higher than 0.05%
```

```python
## Note : We reject Null hypothesis only if pvalue<0.05%
```

```python
## Solution :We cannot accept Null hypthesis since the pvalue is greater in this case
#3 There is a significant difference. Proceed with Alternative Hypothesis
```
