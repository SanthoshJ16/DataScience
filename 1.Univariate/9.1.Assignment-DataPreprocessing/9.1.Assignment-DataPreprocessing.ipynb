```python
## USe Imputers to fill the NAN column data
import numpy as np
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
dataset[quan] = imputer.fit_transform(dataset[quan])

## Make sure to fill the empty cells with NAN before using the Imputer logic
dataset[qual] = dataset[qual].replace(['?', ' ?', '', '  '],np.nan)

```

```python

# compute mode fol all columns
mode = dataset[qual].mode().iloc[0]

# fill NaN values using mode
dataset[qual] = dataset[qual].fillna(mode)
```
