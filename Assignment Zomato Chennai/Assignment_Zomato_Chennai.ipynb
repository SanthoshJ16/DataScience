```python
def quanQual(dataset):
    qual=[]
    quan=[]
    for columnName in dataset.columns:
        if(dataset[columnName].dtype=="O"):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan, qual
```

###### Cleaning the records for Dining rating and Delivery Rating

```python
## Cleaning the records for Dining rating and Delivery Rating
## Step 1: Replacing the is null with 0

dataset.fillna({'Dining Rating': 0}, inplace=True)
dataset.fillna({'Delivery Rating': 0}, inplace=True)
```

```python
## Step2: Removing the useless data from Dining rating and Delivery Rating

# Finding those restaurant whose has 0 Dining rating  or Votes
useless_data = (dataset['Dining Rating'] == 0.0) | (dataset['Delivery Rating'] == 0)
dataset[useless_data]
```

```python
## Performing Negation of the useless dataset and then storing the correct data back in the raw_df DataFrame
## This permamnently remove the wrong data from the original dataframe
dataset = dataset[~useless_data]
```

```python
## Confirming that no Null records in DAYS_OPEN column
dataset.isnull().sum()
```

##### Which restaurants perform significantly better in delivery than dining 

```python
# To identify restaurants that perform significantly better in delivery than dining, 
# we compare Delivery Rating vs Dining Rating at the restaurant level
## Step 1 Finding the Aggregate dining & delivery ratings per restaurant

ratings_comparison = (dataset.groupby('Name of Restaurant').agg(avg_dining_rating=('Dining Rating', 'mean'),
                    avg_delivery_rating=('Delivery Rating', 'mean'), dining_count=('Dining Rating', 'count'),
                    delivery_count=('Delivery Rating', 'count')))
```

```python
# Step 2: Compute the performance gap between Delivery & Dining
ratings_comparison['rating_diff'] = (ratings_comparison['avg_delivery_rating']
                                     - ratings_comparison['avg_dining_rating'])
```

```python
# Step 3: Filter the restaurants that perform significantly better in delivery
better_in_delivery = (ratings_comparison.query('rating_diff >= 0.5 and delivery_count >= 20 and dining_count >= 20')
                        .sort_values('rating_diff', ascending=False))
```

```python
# Step 4 View the top 10 performing restaurants
better_in_delivery.head(10)
```

```python
### Step 5: Plotting in Bar chart
top_delivery_better = better_in_delivery.head(10)


plt.figure(figsize=(12, 6))

colors = plt.cm.tab20(np.linspace(0, 1, len(top_delivery_better)))
bars = plt.barh(top_delivery_better.index,top_delivery_better['rating_diff'],color=colors)

plt.xlabel('Delivery Rating − Dining Rating')
plt.ylabel('Restaurant')
plt.title('Restaurants Performing Better in Delivery than Dining')

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.02,bar.get_y() + bar.get_height()/2,f'{width:.2f}',va='center')

plt.tight_layout()
plt.show()
```

##### Finding the top 10 cuisines in chennai based on price , dining , dining rating, Delivery , Delivery Rating

```python
## Step 1: Splitting the restaurant and cuisine
df = dataset.copy()
df['Cuisine'] = df['Cuisine'].str.split(', ')
df = df.explode('Cuisine')
```

```python
## Step 2: Aggregate metrics at cuisine level
cuisine_metrics = (df.groupby('Cuisine').agg(
        avg_price=('Price for 2', 'mean'),
        avg_dining_rating=('Dining Rating', 'mean'),
        dining_popularity=('Dining Rating Count', 'mean'),
        avg_delivery_rating=('Delivery Rating', 'mean'),
        delivery_popularity=('Delivery Rating Count', 'mean'),
        restaurant_count=('Name of Restaurant', 'nunique')))
```

```python
# Step 3: Remove cuisines with very few restaurants
cuisine_metrics = cuisine_metrics[cuisine_metrics['restaurant_count'] >= 10]
```

```python
## Step 4: Combine the overall score
cuisine_metrics['overall_score'] = ( cuisine_metrics['avg_dining_rating'] +
                                     cuisine_metrics['avg_delivery_rating'])
```

```python
### Step 5: Find top 10 cuisines
top_10_cuisines = (cuisine_metrics.sort_values('overall_score', ascending=False).head(10))

top_10_cuisines
```

```python
### Step 6: Plotting the graph Based on Dining vs Delivery Ratings for top 10 Cuisines

top_10_cuisines[['avg_dining_rating', 'avg_delivery_rating']].plot(kind='bar',figsize=(14, 6))

plt.title('Top 10 Cuisines – Dining vs Delivery Ratings')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
